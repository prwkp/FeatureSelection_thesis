```python
def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
# for threshold
def filter_arr(sample_arr, filter_index):
    result_val = sample_arr[filter_index]
    result_index = filter_index
    compare_index = filter_index + 1
    if sample_arr[0] == sample_arr[filter_index]:
        return (0,0)
        
    if (len(sample_arr) -1 == filter_index):
        compare_index = filter_index - 1
    
    if (sample_arr[filter_index] == sample_arr[compare_index]):
        for index, val in enumerate(sample_arr):
            if (val == result_val):
                result_val = sample_arr[index - 1]
                result_index = index - 1

    return (result_val, result_index)

```

```python
def crossvalidresults(model,X,y):
    rfc_cv_score = cross_val_score(model, X, y, cv=10)
    rfc_cv_f1 = cross_val_score(model, X, y, scoring="f1", cv = 10).mean()
    rfc_cv_f1sd = cross_val_score(model, X, y, scoring="f1", cv = 10).std()
    print("=== Mean F1 Score ===")
    print("Mean F1 Score - : ", rfc_cv_f1)
    print("=== sd F1 Score ===")
    print("sd F1 Score - : ", rfc_cv_f1sd)
    print('\n')
```

```python
output_single = dict()
start_time = timer(None)

for tres,t in thres_d.items():
    for met,m in method.items():
        k =(filter_arr(m.filter(regex='value').squeeze(), t-1))   
        if k[1] != 0:
            n = k[1]+1 
            print("n"+"_single"+met+"_"+tres+"XGBoost",n)
            output_single["n"+"_single"+met+"_"+tres+"XGBoost"] = n
            df_selected = m.iloc[:k[1]+1]
            df_filter = df1.filter(items=df_selected['Feature'])
            X = df_filter
            y = df1['class']
            
            ## before tuned ##
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123) 
            xgbc = XGBClassifier(random_state=123)
            xgbc.fit(X_train,y_train)
            ## results F1 and stdF1 ##
            m = cross_val_score(xgbc, X, y, scoring="f1", cv = 10).mean() 
            s = cross_val_score(xgbc, X, y, scoring="f1", cv = 10).std() 
            print("avgF1"+"_single"+met+"_"+tres+"XGBoost" , m)
            print("stdF1"+"_single"+met+"_"+tres+"XGBoost",s)
            output_single["avgF1"+"_single"+met+"_"+tres+"XGBoost"] = m
            output_single["stdF1"+"_single"+met+"_"+tres+"XGBoost"] = s
            
            for parm, p in param_d.items():
                # Instantiate the grid search model
                grid_search = GridSearchCV(estimator = xgbc, param_grid = p,
                                        cv = 5, n_jobs = -1, verbose = 2,scoring='f1')
                # Fit the grid search to the data
                grid_search.fit(X_train, y_train)
        #         print(grid_search.best_params_)
                xgbc_tuned = XGBClassifier(subsample=grid_search.best_params_['subsample'],
                                                max_depth=grid_search.best_params_['max_depth'],
                                                colsample_bytree=grid_search.best_params_['colsample_bytree'],
                                                min_child_weight=grid_search.best_params_['min_child_weight'],
                                                learning_rate=grid_search.best_params_['learning_rate'],
                                                eta=grid_search.best_params_['eta'])
                xgbc_tuned.fit(X_train,y_train)

                ## results F1 and stdF1 ##
                m_tune  = cross_val_score(xgbc_tuned, X, y, scoring="f1", cv = 10).mean() 
                s_tune  = cross_val_score(xgbc_tuned, X, y, scoring="f1", cv = 10).std() 
                print("avgF1"+parm+"_single"+met+"_"+tres+"XGBoost" , m_tune)
                print("stdF1"+parm+"_single"+met+"_"+tres+"XGBoost",s_tune)
                output_single["avgF1"+parm+"_single"+met+"_"+tres+"XGBoost"] = m_tune
                output_single["stdF1"+parm+"_single"+met+"_"+tres+"XGBoost"] = s_tune

        else:
            output_single["n"+"_single"+met+"_"+tres+"XGBoost"] = k[1]
            output_single["avgF1"+"_single"+met+"_"+tres+"XGBoost"] = k[1]
            output_single["stdF1"+"_single"+met+"_"+tres+"XGBoost"] = k[1]
            for parm, p in param_d.items():
                output_single["avgF1"+parm+"_single"+met+"_"+tres+"XGBoost"] = k[1]
                output_single["stdF1"+parm+"_single"+met+"_"+tres+"XGBoost"] = k[1]
         

        
        
timer(start_time) 
```

```python
mi_XGBoostfeatures = pd.read_csv('mi_XGBoostfeatures.csv')
SHAP_XGBoostfeatures = pd.read_csv('SHAP_XGBoostfeatures.csv')
embed_XGBoostfeatures = pd.read_csv('embed_XGBoostfeatures.csv')
rank = list(range(1,(len(X_.columns)+1)))
            
f_df = pd.DataFrame(
    {
     'mi_XGBoostfeatures': list(mi_XGBoostfeatures['Feature']),
     'SHAP_XGBoostfeatures': list(SHAP_XGBoostfeatures['Feature']),
     'embed_XGBoostfeatures': list(embed_XGBoostfeatures['Feature']),
        'rank_feature': rank
    })

f_df1 = f_df[['mi_XGBoostfeatures','rank_feature']].copy()
f_df2 = f_df[['SHAP_XGBoostfeatures','rank_feature']].copy()
f_df3 = f_df[['embed_XGBoostfeatures','rank_feature']].copy()

f_df12 = pd.merge(left=f_df1[['mi_XGBoostfeatures','rank_feature']],
                     right = f_df2[['SHAP_XGBoostfeatures','rank_feature']], 
                     how = 'left' ,
                     left_on = 'mi_XGBoostfeatures',
                     right_on = 'SHAP_XGBoostfeatures'
                    )

f_df123 = pd.merge(left=f_df12[['mi_XGBoostfeatures','rank_feature_x','SHAP_XGBoostfeatures','rank_feature_y']],
                     right = f_df3[['embed_XGBoostfeatures','rank_feature']], 
                     how = 'left' ,
                     left_on = 'mi_XGBoostfeatures',
                     right_on = 'embed_XGBoostfeatures'
                    )
f_df123 = f_df123.drop(columns = ['SHAP_XGBoostfeatures','embed_XGBoostfeatures'])
f_df123.columns = ['Feature', 'rank_mi', 'rank_SHAP', 'rank_embed']
f_df123 = f_df123.set_index('Feature')

f_df123['value_min'] =f_df123[['rank_mi','rank_SHAP','rank_embed']].min(axis = 1)
f_df123['value_med'] =f_df123[['rank_mi','rank_SHAP','rank_embed']].median(axis = 1)
f_df123['value_mean'] =f_df123[['rank_mi','rank_SHAP','rank_embed']].mean(axis = 1)


d = 1/f_df123['rank_mi']
e = 1/f_df123['rank_SHAP']
f = 1/f_df123['rank_embed']
f_df123['value_harmo'] = 3/(d+e+f)


f_df123.to_csv('f_df123_xgb.csv', index=True)
f_df123
```

```python
def union_3FS(dfMIcut, dfSHAPcut, dfEmbedcut):
    feature_TC = list(set().union(dfMIcut+dfSHAPcut+dfEmbedcut))
    return feature_TC

def multi_3FS(dfMIcut, dfSHAPcut, dfEmbedcut):
    fea12_inters = list(set(dfMIcut).intersection(set(dfSHAPcut)))
    fea13_inters = list(set(dfMIcut).intersection(set(dfEmbedcut)))
    fea23_inters = list(set(dfSHAPcut).intersection(set(dfEmbedcut)))
    fea123_inters = list(set(dfMIcut).intersection(set(dfSHAPcut),set(dfEmbedcut)))
    feature_TC = list(set().union(fea12_inters+fea13_inters+fea23_inters+fea123_inters))
    return feature_TC


def inter_3FS(dfMIcut, dfSHAPcut, dfEmbedcut):
    fea123_inters = list(set(dfMIcut).intersection(set(dfSHAPcut),set(dfEmbedcut)))
    return fea123_inters
```

```python
method_TC = {'union': union_3FS,
'multi':multi_3FS, 
'inters':inter_3FS
}
```
